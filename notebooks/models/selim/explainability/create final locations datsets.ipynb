```python
origins_country_kwords = [
 'burundian',
 'ecuadorian',
 'guatemalan',
 'iranian',
 'malian',
 'nicaraguan',
 'peruvian',
 'syrian',
 'venezuelan',
 "libyan",
 "ukrainian",
 "kenyan",
 "australian",
 "italian",
 "colombian",
 "nigerian",
 "chilian",
 "canadian",
 "american",
 "mexican",
 ]


origins_continent_kwords = ['african', 'european', "asian"]
```

```python
locations_countries = [
 'argentina',
 'bangladesh',
 'brazil,',
 'congo',
 'ecuador',
 'guatemala',
 'honduras',
 'lebanon',
 'mali',
 'myanmar',
 'nicaragua',
 'niger',
 'peru',
 'sudan',
 'syria',
 'uruguay',
 'venezuela', "france", "italy", "switzerland", "canada", "ukraine", "turkey", "chile", "colombia", "libya", "burundi", "poland"]

locations_continents = ['africa', 'europe', "asia"]

locations_cities = ['bogotá', 'damascus', "geneva", "paris", "istanbul", "tokyo", "kiev"]
```

```python
str(locations_continents)
```

```python
def get_augmented_df(original_df, kw: List[str], name: str):

    df = original_df.copy()
    df['kw'] = df['excerpt'].apply(
        lambda x: [word for word in kw if word in x.lower()]
    )
    df = df[df.kw.apply(lambda x: len(x)==1)]
    df['kw'] = df['kw'].apply(lambda x: x[0])
    df['type'] = 'original'
    print(len(df))

    final_df = pd.DataFrame()
    for i in range(df.shape[0]):
        row = df.iloc[[i]]
        original_row_kw = row["kw"].values[0]
        row["type"] = "augmented"
        for one_kw in kw:
            if one_kw != original_row_kw:
                row_tmp = row.copy()
                row_tmp["excerpt"] = row_tmp["excerpt"].apply(
                    lambda x: x.replace(original_row_kw, one_kw).replace(
                        original_row_kw.capitalize(), one_kw.capitalize()
                    )
                )
                row_tmp["kw"] = [one_kw]
                final_df = final_df.append(row_tmp)

    final_df = pd.concat(
        [df, final_df]
    )

    final_df.to_csv(f'final_data/{name}.csv', index=None)

    return final_df
```

```python
augmented_locations_continents = get_augmented_df(
    locations_df[['entry_id', 'excerpt']],
    kw=locations_continents,
    name='locations_continents'
)
display(augmented_locations_continents)

augmented_locations_countries = get_augmented_df(
    locations_df[['entry_id', 'excerpt']],
    kw=locations_countries,
    name='locations_countries'
)
display(augmented_locations_countries)

augmented_locations_cities = get_augmented_df(
    locations_df[['entry_id', 'excerpt']],
    kw=locations_cities,
    name='locations_cities'
)
display(augmented_locations_cities)

augmented_origins_country = get_augmented_df(
    origins_df[['entry_id', 'excerpt']],
    kw=origins_country_kwords,
    name='origins_country'
)
display(augmented_origins_country)

augmented_origins_continent = get_augmented_df(
    origins_df[['entry_id', 'excerpt']],
    kw=origins_continent_kwords,
    name='origins_continent'
)
display(augmented_origins_continent)
```
